# 3. Clean the dataset by removing missing values and other things.

# 4. Transform attributes into numerical values and other necessary transformations.


**Changing float into integers**

```
train_data["views"]=pd.to_numeric(train_data["views"])
train_data["likes"]=pd.to_numeric(train_data["likes"])
train_data["dislikes"]=pd.to_numeric(train_data["dislikes"])
train_data["comment"]=pd.to_numeric(train_data["comment"])
train_data["adview"]=pd.to_numeric(train_data["adview"])
col_vidid=train_data["vidid"]
```

**Encoding the attributes**

```
from sklearn.preprocessing import LabelEncoder
train_data["duration"]=LabelEncoder().fit_transform(train_data["duration"])
train_data["vidid"]=LabelEncoder().fit_transform(train_data["vidid"])
train_data["published"]=LabelEncoder().fit_transform(train_data["published"])
train_data.head(10)
```

**Converting time in seconds**

```
import datetime
import time
def check(x):
  y=x[2:]
  h=''
  m=''
  s=''                     
  time=''
  p=['H','M','S']      #ad duration
  for i in y:
    if i not in p:
      time+=i
    else:
      if (i=="H"):
        h=time
        time=''
      elif (i=="M"):
        m=time
        time=''
      else:
        s=time
        time=''
  if (h==''):
    h='00'
  if (m==''):
    m='00'
  if (s==''):
    s='00'
  t=h+':'+m+':'+s
  return t
train=pd.read_csv("train.csv")
tm=pd.read_csv("train.csv")["duration"]
tot_tim=tm.apply(check)
def tosec(time_string):
  h,m,s =time_string.split(':')
  return int(h) * 3600 + int(m) * 60 + int(s)
time1 = tot_tim.apply(tosec)
train_data["duration"]=time1
train_data.head()
```

# 6. Using Linear Regression, Support vector Regressor for training and getting errors.

**Metrics for evaluation**

```
from sklearn import metrics
def mod_error(x_test,y_test,mod_nm):
  pred=mod_nm.predict(x_test)
  print("Mean Absolute Error: ",metrics.mean_absolute_error(y_test,pred))
  print("Mean Squared Error: ",metrics.mean_squared_error(y_test,pred))
  print("Root Mean Square Error: ",np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```
from sklearn import linear_model
linear_regression=linear_model.LinearRegression()
linear_regression.fit(x_train, y_train)
mod_error(x_test, y_test, linear_regression)
```

```
train_data.describe()
```

```
train_data.isnull().any()
```

```
from sklearn.svm import SVR
supportvector_regressor=SVR()
supportvector_regressor.fit(x_train, y_train)
mod_error(x_test,y_test,supportvector_regressor)
```

# 7. Using Decision Tree Regressor and Random Forest Regressors.

**Decision Tree Regressor**

```
from sklearn.tree import DecisionTreeRegressor
dec_tree=DecisionTreeRegressor()
dec_tree.fit(x_train,y_train)
mod_error(x_test,y_test,dec_tree)
```

**Random Forest Regressor**

```
from sklearn.ensemble import RandomForestRegressor
n_est=200
max_dep=25
minsamsplit=15
minsamleaf=2
rand_for=RandomForestRegressor(n_est,max_depth=max_dep,min_samples_split=minsamsplit,min_samples_leaf=minsamleaf)
rand_for.fit(x_train,y_train)
mod_error(x_test,y_test,rand_for)
```

# 8. Building an artificial neurat network and training it with different layers and hyperparameters using keras.
